```python
# Text analysis of Alice's Adventures in Wonderland

import pandas as pd
import numpy as np
import nltk
import os
import nltk.corpus
with open('/Users/nicolemark/Desktop/alice_full_text.csv') as f:
    text = f.read()
f.close()
```

```python
# importing word_tokenize from nltk

from nltk.tokenize import word_tokenize

# Passing the string text into word tokenize for breaking the sentences
token = word_tokenize(text)
token
```

```python
# Remove punctuation
words = token
new_words= [word for word in words if word.isalnum()]

print(new_words)
```

```python
# Remove stopwords such as "a", "the"

from nltk.corpus import stopwords
a = set(stopwords.words('english'))
text_2 = [x for x in new_words if x not in a]
print(text_2)
```

```python
text_2
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(text_2)
fdist
```

```python
fdist1 = fdist.most_common(1000)
fdist1
```

```python
# Create DataFrame containing most common words and their frequencies. I want the frequency of all words, so I adjusted the code for fdist accordingly.

df = pd.DataFrame(fdist1)
df
```

```python
from matplotlib import pyplot as plt
df.hist()
```

```python
# Add column names in place of indices
df.rename(columns={0: 'Word', 1: 'Frequency'}, inplace=True)
```

```python
df.to_csv('alice_word_freq.csv', encoding='utf-8', index=False)
```
